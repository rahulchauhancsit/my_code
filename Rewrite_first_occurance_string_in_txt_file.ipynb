```python
import os, sys
import glob
import fileinput

root_dir = "E:\Vaaan\Working\Data_1\Vehicles"


for filename in glob.iglob(root_dir + '**/*.txt', recursive=True):
    print(filename)
    # replace all occurrences of '36' with '0' and insert a line after the 1th
    for i, line in enumerate(fileinput.input(filename)):
        sys.stdout.write(line.replace('36', '0', 1)) 
        f = open(filename,"w")
        f.write(line.replace('36', '0', 1))
        f.close
        # replace '36' and write '0'
    if i == 0: 
        sys.stdout.write('\n')  
```
